```python
# as id in meta is object type. converting that to int
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
data.head()
```

```python
# we have all the movie information in the columns genres, cast and crew. we need to extract only required information
# from these features
```

```python
# genres and cast have informaiton in the form list of dictionary: that is key value pairs in a list and in the form of string
# using the package ast to extract information
```

```python
import ast

# the ast.literal_eval() function converts the list of dictionaries to list

data['genres'] = data['genres'].apply(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].apply(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].apply(lambda x: ast.literal_eval(x))
```

```python
data['genres'][1]
```

```python
# defining a function to get only the names of the genres from the genre feature:
def make_genreList(x):
    gen = []
    st = " " 
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NAN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].apply(lambda x: make_genreList(x))
```

```python
data['genres_list']
```

```python
# defining a function to get only the names of the actor 1 from the cast feature:
```

```python
data['cast'][1]
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        actor_name = i.get('name')
        casts.append(actor_name)
    if casts == []:
        return np.NAN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data['actor_1_name']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        actor_name = i.get('name')
        casts.append(actor_name)
    if casts == [] or len(casts) <= 1:
        return np.NAN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        actor_name = i.get('name')
        casts.append(actor_name)
    if casts == [] or len(casts) <= 2:
        return np.NAN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['crew'][1]
```

```python
def get_directors(x):
    director = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            director_name = i.get('name')
            director.append(director_name)
    if director == []:
        return np.NaN
    else:
        return (st.join(director))
```

```python
data['director_name']= data['crew'].apply(lambda x: get_directors(x))
```
